```python

vars = train.select_dtypes(include ='object').columns
nunique = train[vars].nunique()
nunique = nunique.drop('target')
vars = nunique[(train[vars].nunique()>1)&(train[vars].nunique()<10)].index
target = "target"


for i, var in enumerate(vars):
    
    plt.figure(figsize=(10, 10))
    plt.subplot(4,2,i+1) 
    # crosstab을 사용하여 분포 계산
    var_counts = pd.crosstab(train[var], train[target])
    var_counts.plot(kind="bar", stacked=False, color=["red", "blue"], alpha=0.5, ax=plt.gca())

    plt.title(f"{var} Distribution by {target}")
    plt.xlabel(var)
    plt.ylabel(target)
    plt.legend(title=target, labels=["Normal", "AbNormal"])
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

## 데이터 전처리

```python
# target 값 분리
target = train.pop('target')
```

```python
# 'AbNormal': 0, 'Normal': 1
target = target.replace({'AbNormal': 0, 'Normal': 1})
target.value_counts(normalize=True)
```

```python
# # object 컬럼 삭제
cols = train.select_dtypes(exclude='O').columns
# train = train[cols]
# test = test[cols]
```

```python
cols = cols.tolist()
vars = vars.tolist()
```

```python
cols = vars+cols
cols
```

```python
train = train[cols]
test = test[cols]
train
```

```python
vars
```

```python
from sklearn.preprocessing import LabelEncoder
for var in vars:
    le = LabelEncoder()
    train[var] = le.fit_transform(train[var])
    test[var] = le.transform(test[var])
```

```python
# StandatdScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
train = scaler.fit_transform(train)
test = scaler.transform(test)
```

```python
# train_test_split
from sklearn.model_selection import train_test_split
X_tr, X_val, y_tr, y_val = train_test_split(train, target, test_size=0.2, random_state=2024)
X_tr.shape, X_val.shape, y_tr.shape, y_val.shape
```

## 모형 학습 밎 검증

```python
# 평가 라이브러리(accuracy_score, F-1 score, roc_auc_score)
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score
```

```python
# LinearRegression
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_tr,y_tr)
y_predict = lr.predict(X_val)
y_pred = (y_predict>0.5).astype(int)
# print(y_pred)

# 평가
print(f1_score(y_val,y_pred))
print(roc_auc_score(y_val, y_predict))
print(accuracy_score(y_val, y_pred))
```

```python
#LogisticRegression
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(X_tr,y_tr)
y_predict = log.predict(X_val)
y_pred = log.predict_proba(X_val)[:,1]

# 평가
print(f1_score(y_val,y_predict))
print(roc_auc_score(y_val, y_pred ))
print(accuracy_score(y_val, y_predict))
```

```python
#DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(max_depth=5, min_samples_split=2, min_impurity_decrease=0.0)
dtc.fit(X_tr, y_tr)
y_predict = dtc.predict(X_val)
y_pred = dtc.predict_proba(X_val)[:,1]

# 평가
print(f1_score(y_val, y_predict))
print(roc_auc_score(y_val, y_pred ))
print(accuracy_score(y_val, y_predict))
```

```python
# Bagging
from sklearn.ensemble import BaggingClassifier
base_model = DecisionTreeClassifier()
bag = BaggingClassifier(estimator=base_model,
                          n_estimators=300,
                          bootstrap=True,
                          n_jobs=-1,
                          random_state=110)
bag.fit(X_tr,y_tr)
y_predict = bag.predict(X_val)
y_pred = bag.predict_proba(X_val)[:,1]

# 평가
print(f1_score(y_val, y_predict))
print(roc_auc_score(y_val, y_pred ))
print(accuracy_score(y_val, y_predict))
```

```python
# RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(max_depth=9, min_samples_split=2, bootstrap=True,n_jobs=-1, random_state=110)
rf.fit(X_tr,y_tr)

y_predict = rf.predict(X_val)
y_pred = rf.predict_proba(X_val)[:,1]

# 평가
print(f1_score(y_val, y_predict))
print(roc_auc_score(y_val, y_pred ))
print(accuracy_score(y_val, y_predict))
```

```python
#AdaBoost
from sklearn.ensemble import AdaBoostClassifier
ada = AdaBoostClassifier(n_estimators=100, random_state=110)
ada.fit(X_tr,y_tr)

y_predict = ada.predict(X_val)
y_pred = ada.predict_proba(X_val)[:,1]

# 평가
print(f1_score(y_val, y_predict))
print(roc_auc_score(y_val, y_pred ))
print(accuracy_score(y_val, y_predict))
```

```python
# GradientBoosting
from sklearn.ensemble import GradientBoostingClassifier
gb = GradientBoostingClassifier(n_estimators=100)
gb.fit(X_tr,y_tr)

y_predict = gb.predict(X_val)
y_pred = gb.predict_proba(X_val)[:,1]

# 평가
print(f1_score(y_val, y_predict))
print(roc_auc_score(y_val, y_pred ))
print(accuracy_score(y_val, y_predict))
```
